### Linearizing Functions
Why is the line equation so useful? The main reason is because it's so easy to work with. Scientists actually try their best to linearize functions, that is, to make linear functions from non-linear ones. There are several ways of doing this. One of them involves derivatives and we'll talk about it later in the course. 

A commonly used method for linearizing functions is through algebraic transformations. Try to linearize the exponential function
$$ y = ae^{bx} $$
where $e$ is _the base of the natural logarithm_.

**Hint:** The inverse operation of $e^{x}$ is $\ln(x)$. Start by taking $\ln$ of both sides and see what you can do. Your goal is to transform the function into another, linear function. You can look up more hints on the Internet if you get too stuck :).

$$ y = ae^{bx} $$

Now let's apply **logarithmic transformation** (on both sides of the equation).

$$ ln(y) = ln(ae^{bx}) $$

After that we apply the **"Product Rule"** of the log properties.

$$ ln(y) = ln(a) + ln(e^{bx}) $$

Now we apply the **"Power Rule"** of the log properties.

$$ ln(y) = ln(a) + bx \cdot ln(e) $$

And finally we acknowledge that $ln(e)$ is equal to 1.

$$ ln(y) = ln(a) + bx $$

Now we have a linear function! Congratulations!

$$ ln(y) = bx + ln(a) $$


```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def draw_exponential_function(a: float, b: float):
    x = np.linspace(-3, 3, 100)
    y = a * math.e ** (b*x)

    # Stylize the plot.
    ax = plt.gca()
    ax.spines["bottom"].set_position("zero")
    ax.spines["left"].set_position("zero")
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    # Draw plot
    plt.plot(x, y)
    plt.show()

draw_exponential_function(5, 2) # Randomly pick values for "a" and "b"
```

```python
def draw_linear_function(a: float, b: float):
    x = np.linspace(-3, 3, 100)
    ln_y = b * x + math.log(a)

    # Stylize the plot.
    ax = plt.gca()
    ax.spines["bottom"].set_position("zero")
    ax.spines["left"].set_position("zero")
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)

    # Draw plot
    plt.plot(x, ln_y)
    plt.show()

draw_linear_function(5, 2) # Randomly pick values for "a" and "b"
```
