````python
import math
import matplotlib.pyplot as plt
````

````python
# Q1
# Domain = all the values of the INPUT params (What can go into a function is called "Domain")
# Codomain = all the possible OUTPUT values of a function (What may possibly come out of a function is called "Codomain")
# Range = what actually comes OUT of the function (i.e. a subset of the "Codomain")

# Q2 (sqrt)
x_domain = range(0, 100, 1) # Only POSITIVE numbers (and 0) are accepted.
f = math.sqrt
y_range = [f(x) for x in x_domain]
plt.plot(x_domain, y_range)
plt.show()
````

````python
# Q2 (power)
x_domain = range(0, 100, 1)
y_range = [math.pow(x, 2) for x in x_domain]
plt.plot(x_domain, y_range)
plt.show()
````

Visually the ```square root``` and ```power``` looks like "inverse" functions. But let's verify this hypothesis by applying both function on x. If they are are indeed inverse, then we would expect to observe a simple 
$$ y=x $$
drawing.

````python
x_domain = range(0, 100, 1)
y_range = [math.sqrt(math.pow(x, 2)) for x in x_domain]
plt.plot(x_domain, y_range)
plt.show()
````

Indeed, the hypothesis is validated. **Wohoo! My first hypothesis validation was just executed.** :) 

````python
# Simply curious to draw the power function for negative numbers as well.
x_domain = range(-100, 100, 1)
y_range = [math.pow(x, 2) for x in x_domain]
plt.plot(x_domain, y_range)
plt.show()
````

````python
# Q3 Now let's compare all the 3 different variants for powering a number
x_domain = [float("NaN"), -math.inf, -15j, -10.1, -0.0001, 0, 1e-4, 8]
for x in x_domain:
    if type(x) is complex:
        print(f"{x**2}, {pow(x, 2)} |  {x**0}, {pow(x, 0)}")
    else:
        print(f"{x**2}, {pow(x, 2)}, {math.pow(x, 2)} |  {x**0}, {pow(x, 0)}, {math.pow(x, 0)}")
````

Obviously, one of the differences is that ```math.pow()``` always casts the numbers to ```float```. Hence, it does not support ```complex``` numbers.

Now let's delve into **logarithms** and see how can we represent them in an arbitrary base.
$$ 2^x = 16  \Leftrightarrow  x = 4$$
$$ log_2 16 = x \Leftrightarrow  x = 4$$

Or generally put:
$$ log_b a = x \Leftrightarrow a = b^x $$


Now, let's implement logarithmic function in Python. To do so we need to exploit the "[Change of Base rule](https://www.cuemath.com/algebra/properties-of-logarithms/)":
$$ log_b a = \frac{log_k a}{log_k b} $$

Furthermore, we need to use approximate the *Taylor series* of the natural logarithm and come up with following equation:
$$ ln(x) = \lim_{n \rightarrow \infty}{n (x^{1/n} - 1)}$$

````python
# Q4 The implementation return value which is near to the correct answer, but not exact.
def log(base: float, a: float):
    return ln(a)/ln(base)

def ln(a: float):
    return 99999999*(a**(1/99999999) - 1)

log(3, 81)
````
